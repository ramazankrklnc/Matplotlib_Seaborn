## Colormap ekleme

```python
grouped_data = data.groupby(["species"])["body_mass_g"].mean()
grouped_data
```

```python
grouped_data.index
```

```python
sns.barplot(x = grouped_data.index,
            y = grouped_data.values)
'''sns.barplot(data = grouped_data,
               x = "species",
               y = "body_mass_g") '''
# Burda 2. yazdığımız "reset.index" için kullanılan kod. Fakat biz data parametresi vermeden bir arrayle de bu fonksiyonları kullanabiliyoruz.
```

```python
cmap = plt.get_cmap("viridis")
cmap
```

```python
norm = plt.Normalize(grouped_data.min(), grouped_data.max())
norm
```

```python
colors = cmap(norm(grouped_data.values))
colors
```

```python
fig, ax = plt.subplots(figsize = (12, 9))

sns.barplot(x=grouped_data.index,
            y=grouped_data.values,
            palette=colors)              # Viridis'e göre renk değiştirdik.

ax.set_ylabel("Average body_mass_g")     # y'nin etiketi yok ona burdan isim verdik.

cax = fig.add_axes([0.95, 0.15, 0.03, 0.67])        # cax = Color Map axes
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)    # cm = colormap , sm = ScalerMap
fig.colorbar(sm, cax=cax)

plt.show()
```

Kısaca Özeteleyecek Olursak: Bir tane colormap seçiyoruz, bu değerleri normalize ediyoruz 0-1 arasında ve "colors = cmap(norm(grouped_data.values))" şöyle ikili zincieleme bir kol(iç içe parantez yazmak) yaparak yani colabel object yapıyoruz. Colabel object bize geri değerler döndürüyor. Ve burdan colorsları alıcaz ve daha sonra renkleri mapleyebilmek için yardımcı olan fonksiyonu cm (colormap) modülünün altında "ScalarMappable" yardımcı fonksiyonu kullanıp mape alıyoruz. Ve bu mapi de "fig.colorbar" ile matplotlibe veriyoruz ki bunu görselleştirsin. " fig.add_axes" dediğimiz şey ise bu color mapin konumu, genişliğini, büyüklüğünü kontrol eden bir fonksiyondur. Fig konumunu kontrol etmek için de direk olarak plt figureden yararlanmak üzere birebirlik bir splot oluşturduk.
